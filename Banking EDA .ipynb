```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# From numerical data we want to convert Estimated Income to categorical(group then 0-2000, 2000-4000 like that)
'''pd.cut() takes the column Estimated Income and:
   Divides it into bins using the ranges defined
   Labels them using labels
   right=False means the intervals are left-inclusive:
   So, 100000 goes into 'Med' (not 'Low'), because 'Low' ends at 100000 exclusive'''

bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
# Gives the count of values in that interval 
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
categorical_cols = df[["BRId", "GenderId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "IAId", "Income Band"]].columns
for col in categorical_cols:
    print(f"Value counts for {col}:")
    display(df[col].value_counts())
```

```python
# For univariate analysis same code can be used for bivariate analysis just add hue='colname' in last line 
'''means enumerate stores the number while predictor has the actual column name for that number given by enumerate'''
for i, predictor in enumerate(df[["BRId", "GenderId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "IAId", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
# Bivariate Analysis 
for i, predictor in enumerate(df[["BRId", "GenderId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "IAId", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize = (8, 4))
    sns.histplot(df[col])
    plt.title('Histogram')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Numerical columns analysis 
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate Analysis and visualizations
plt.figure(figsize=(8, 9))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

```python
# Heatmaps
'''correlation_matrix = df[numerical_cols].corr() => This calculates the correlation between each pair of columns.
                                                     It returns values from -1 to 1:
                                                     +1 means a strong positive relationship (they rise together)
                                                     -1 means a strong negative relationship (one rises, the other falls)
                                                     0 means no linear relationship
 
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f") => sns.heatmap() plots the matrix visually as a color-coded grid.
                                                                        annot=True shows the actual correlation values inside the boxes.
                                                                        cmap='crest' gives a nice blue-green color palette.
                                                                        fmt=".2f" limits the numbers to two decimal places for clarity.

'''

numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize = (12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
